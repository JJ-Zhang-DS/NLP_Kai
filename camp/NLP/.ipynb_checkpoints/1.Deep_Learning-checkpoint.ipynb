### Perceptron/Neuron (1957 by Frank Rosenblatt)
This is the most fundamental building block of neural network. It is basically a linear tranformer, meaning every output is a linear combination of input.

### Max Pooling and Average Pooling: 
The purpose of the pooling layer is to reduce the size of the image. The whole neural network is basically an information distillation system, and pooling layer is to reduce the spatial information while keeping the pattern. 

### Activation Function: 
A perceptron is a linear operator, which means the output will scale linearly with the input. But a lot of phenomenons are non-linear. The resistance disappear when temperature drops to certain point for super-conductors; a neuron is activated when the signal from its neighbor is high enough. Activation layer does the same thing. Besides, activation function can serve to reduce overfitting by regularizing the input.

There are tons of different activation functions. As long as it brings non-linearity, you can invent any function form. The most popular ones are Sigmoid and ReLU. Look at their curves and see how non-linearity happens.

## Deep learning problems
The idea of deep learning is simple: stack many layers together and make the neural network deep. But, why deep neural network works so well? Why it out-performs all previous models like SVM and tree models? 

Part of the answer lies in the fact that: even a simple one-layer shallow neural network can fit any functions. Refer to for an explanation here: http://neuralnetworksanddeeplearning.com/chap4.html . By comparison, linear SVM tries to find the linear separation hyper-plane. The shape of the separation plane can be bended using kernel trick, but not as flexible as neural network. 

But this is just a theoretical result, how can we really fit the function we need? We will need a loss function to quantify how good the fitting is, and a smart way to find the minimal value of loss function.

Unfortunately, the loss function of neural network is not convex.

## 1. <a name="1">Implementing a neural network with PyTorch</a>
(<a href="#s1">Go to top</a>)

Let's implement a simple neural network with two hidden layers of size 64 using the sequential container (Adding things in sequence). We will have 3 inputs, 2 hidden layers and 1 output layer.

````python
import torch
from torch import nn

net = nn.Sequential(
        nn.Linear(3, 64),     # Linear layer-1 with 64 out_features and input size 3
        nn.Tanh(),            # Tanh activation is applied
        nn.Linear(64, 64),    # Linear layer-2 with 64 units and input size 64  
        nn.Tanh(),            # Tanh activation is applied
        nn.Linear(64, 1))     # Output layer with single unit

print(net)
````

The weight parameters of the `Linear` layer in pytorch are initialized with a modified form of the Xavier Initialization. Using these weights as a start, we can later apply optimization such as SGD to train the weights. As a result, using a strategic technique to initialize the weights is crucial.

## 4. <a name="4">Example - Binary Classification</a>
(<a href="#s1">Go to top</a>)

In this example, we will train a neural network on a dataset that we randomly generated. We will have two classes and train a neural network to classify them.

````python
from sklearn.datasets import make_circles

X, y = make_circles(n_samples=750, shuffle=True, random_state=42, noise=0.05, factor=0.3)
````

````python
X # dots on 2D
````

````python
y[:10] # 0: inner, 1: outer
````

````python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns

def plot_dataset(X, y, title):
    
    # Activate Seaborn visualization
    sns.set()
    
    # Plot both classes: Class1->Blue, Class2->Red
    plt.scatter(X[y==1, 0], X[y==1, 1], c='blue', label="class 1")
    plt.scatter(X[y==0, 0], X[y==0, 1], c='red', label="class 2")
    plt.legend(loc='upper right')
    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.xlim(-2, 2)
    plt.ylim(-2, 2)
    plt.title(title)
    plt.show()
    
plot_dataset(X, y, title="Dataset")
````

````python
import time
from torch.nn import BCELoss
````

Then, we create the network as below. It will have two hidden layers. Since the data seems easily seperable, we can have a small network (2 hidden layers) with 10 units at each layer.

````python
# Use GPU resource if available, otherwise wil use CPU
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")

net = nn.Sequential(nn.Linear(in_features=2, out_features=10),
                    nn.ReLU(),
                    nn.Linear(10, 10),
                    nn.ReLU(),
                    nn.Linear(10, 1),  #binaryoutput1label
                    nn.Sigmoid()).to(device)
````

Now let's prepare the training set and validation set, and load each of them to a `DataLoader`, respectively.

````python
device
````

````python
# Split the dataset into two parts: 80%-20% split
X_train, X_val = X[0:int(len(X)*0.8), :], X[int(len(X)*0.8):, :]
y_train, y_val = y[:int(len(X)*0.8)], y[int(len(X)*0.8):]

# Use PyTorch DataLoaders to load the data in batches
batch_size = 4           # How many samples to use for each weight update 
train_dataset = torch.utils.data.TensorDataset(torch.tensor(X_train, dtype=torch.float32),
                                               torch.tensor(y_train, dtype=torch.float32))
train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=batch_size)

# Move validation dataset on CPU/GPU device
X_val = torch.tensor(X_val, dtype=torch.float32).to(device)
y_val = torch.tensor(y_val, dtype=torch.float32).to(device)
````

````python
epochs = 50   # Total number of iterations
lr = 0.01     # Learning rate

# Define the loss. As we used sigmoid in the last layer, we use `nn.BCELoss`.
# Otherwise we could have made use of `nn.BCEWithLogitsLoss`. sigmoidnn.BCEWithLogitsLoss
loss = BCELoss(reduction='none') #

# Define the optimizer, SGD with learning rate
optimizer = torch.optim.SGD(net.parameters(), lr=lr)
````

````python
train_losses = []
val_losses = []
for epoch in range(epochs):
    start = time.time()
    training_loss = 0
    # Build a training loop, to train the network
    for idx, (data, target) in enumerate(train_loader):
        # zero the parameter gradients
        optimizer.zero_grad()
        # Put data and label to the correct device
        data = data.to(device) #
        # view(-1, 1) reshapes (batch_size, ) to (batch_size, 1)
        target = target.to(device).view(-1, 1) 
        # Make forward pass
        output = net(data)
        # Calculate loss
        L = loss(output, target).sum() #loss = BCELoss(reduction='sum')
        training_loss += L.item() #item()tensorscalar
        # Make backwards pass (calculate gradients)
        L.backward()
        # Update weights
        optimizer.step()
    
    # Get validation predictions
    val_predictions = net(X_val)
    # Calculate the validation loss
    val_loss = torch.sum(loss(val_predictions, y_val.view(-1, 1))).item()
    
    # Take the average losses
    training_loss = training_loss / len(y_train)
    val_loss = val_loss / len(y_val)
    
    train_losses.append(training_loss)
    val_losses.append(val_loss)
    
    end = time.time()
    # Print the losses every 10 epochs
    if (epoch == 0) or ((epoch+1)%10 == 0):
        print("Epoch %s. Train_loss %f Validation_loss %f Seconds %f" % \
              (epoch, training_loss, val_loss, end-start))
````

````python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns

plt.plot(train_losses, label="Training Loss")
plt.plot(val_losses, label="Validation Loss")
plt.title("Loss values")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.show()
````

````python
%matplotlib inline
import matplotlib.pyplot as plt
from PIL import Image
from matplotlib.image import imread
import torch
import torchvision
from torchvision import transforms
from torchvision.models import resnet50
````

## 1. <a name="1">Downloading a Pretrained Model</a>
(<a href="#s2">Go to top</a>)

With torchvision, we will start with a ResNet 50 neural net trained on ImageNet dataset as our base model. By specifying
`pretrained=True`, it will automatically download the model from the model
zoo if necessary. For more pretrained models, please refer to [torchvision Model Zoo](https://pytorch.org/vision/stable/models.html).

````python
net = resnet50(pretrained=True)
# weights: The pretrained weights to use. By default, no pre-trained weights are used.
# "pretrained=True" is is equivalent to passing "weights=ResNet50_Weights.IMAGENET1K_V1"
# if using "weights=ResNet50_Weights.DEFAULT" will get the most up-to-date weights
````

## 2. <a name="2">Pre-processing an Image</a>
(<a href="#s2">Go to top</a>)

Next we read a sample image, and pre-process it with preset data transforms `transform_eval`.

````python
img_raw = Image.open('data/catdog.jpg').convert('RGB')
plt.imshow(img_raw)
plt.show()

# transforms.Compose([]) creates a pipeline of transformations to image
transform_eval = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),  # converts the resized image into a PyTorch tensor
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                         std=[0.229, 0.224, 0.225])
]) 
# mean and std values used here are precomputed values that 
#     are commonly used for normalization in computer vision tasks. RGB

img = transform_eval(img_raw)
````

## 3. <a name="3">Inference Using ResNet50</a>
(<a href="#s2">Go to top</a>)

Now let's generate the predictions from the pretrained ResNet50.
`pred` will be a list of ndarray, where each ndarray is of length 1000.
Each number of this 1000-length ndarray can be applied `softmax` to 
represent the prediction confidence towards each of the [subject classes in ImageNet](https://image-net.org/challenges/LSVRC/index.php).

````python
net.eval()  # not in train mode, don't update weights
pred = net(img.unsqueeze(0)) # add a dimension for batch_size
len(pred[0])
````

````python
pred.shape
````

````python
pred.squeeze(0).shape # remove dimension 0 (batch_size)
````

````python
pred.max(), pred.min()
# The output of a ResNet50 model is not probabilities, but rather raw scores before applying softmax 
````

````python
import json

with open("data/imagenet_idx_to_class.json", "r") as f:
    classes = json.load(f)
    
classes[:5]
````

````python
topK = 5
conf, ind = torch.topk(pred.squeeze(0), k=topK)
ind = ind.squeeze(0).numpy()
print('The input picture is classified to be')
for i in range(topK):
    print('\t[%s], with probability %.3f.'%
          (classes[ind[i]], torch.softmax(pred.squeeze(0), dim=0)[ind[i]].item()))
````
