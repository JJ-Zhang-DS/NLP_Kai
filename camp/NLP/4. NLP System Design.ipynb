## Search Engine

As stated in Notebook1, if you know TFIDF in 1990, you can open a startup in Sillicon Valley or Zhongguancun to run search business. The technology behind search engine is simple: vectorize the whole World Wide Web, vectorize each document, and store them somewhere.  When the client input a query, vectorize the query, and find the MOST RELEVANT docs from the database. Then, you can run your business! 

We have done this in Notebook2, trying to retrive top 5 most similar reviews to one review. This is a search system already. The base doc is the query, and you retrieve top 5 most relevant docs from a database of 50K docs. 

This is called <font color='blue'>Information Retrieval</font>. You will see this phrase extremely often. 

But Google's database is not 50K, but much much larger. It is estimated there are over 4trillions of tokens on the WWW. Train a TFIDF model based on this dataset, vectorize each doc, and do the similarity calculation is extremly expensive, thus do not scale. We need some simpler and faster matching algorithms.

Modern data search engines uses approximate Nearest Neighbor (ANN) to do this.

## Recommendation System
<a name="2">General Introduction</a>
(<a href="#0">Go to top</a>)

Recommendation System is fundamentally an Information Retrieval system. Let me explain why. 

### Brutal Force

Think about a brutal force collaborative filtering method for user-based or item-based recommendations. 

User-based method is that you find top k users most similar to your target user, and average their behavior (rating, purchase, like/dislke). Then you can predict what are the most likely items he/she will purchase/like.

Item-based method is reverted. You find top k items that most similar to your target item, and average their relavent history (rating, purchase, like/dislike). Then you can pick top n people who are most likely to engage in the interaction (purchase, like, watch) and do your marketing to them. 

Sounds so similar to search engine, right? Similarity, top k... Because this is fundamentally the same matching game of two sides. 

### Matrix factorization

```python
import utils
utils.show_img('../NLP/img/mf.png', size_inch=18)
```

The link of retrieval system with matrix factorization method is less straightforward, but let me explain briefly. You have a matrix containing the interaction (rating, purchase, like) between customers and items (movie, iPhone etc). Some elements in the matrix is missing, and we are supposed to fill those missing values. To achieve this, you can approximate the interacton matrix with a multiplication of two matrices: user matrix and item matrix. And the full matrix prediction can be obtained by multiplication of the two matrices. 

As shown in the plot above: each user is represented by one vector, and one item is represented by a vector too (The vectors of same color). The dot product of the two vectors give you the interaction prediction. The length of vector: k is different from the number of user: m and number of item: n. In fact k is a hyper-parameter you can adjust. The k variables are called 'latent variable' because they are hidden and not explicitly expressed. They might have some correlations with the features of the users or items, but we don't express it here. 

Recall dot product is used to calculate similarity in previous paragraph, so this is again a similarity based retrieval system! You create embeddings for user and item, and their similarity will tell you how much they are going to interact with each other!

### Transforming recommendation problem into a regression or classification problem

By answering the above questions you probably have realized, <font color='blue'>dealing with interaction matrix is powerful but limited.</font> The latent parameters carry some information of users and items, but why we don't use user and item features directly? Besides, we have a cold start problem that when a new user comes or a new item is launched, they don't have much interaction data and they don't fit into the current interaction matrix. Thus we need to wait for them to accumulate some data and re-train the model. How can we use the features of users and items and solve the cold start problem at the same time?

We can <font color='blue'>convert the recommendation problem into a regression or classification problem by using features to predict the score or interaction.</font> This design logic is super powerful and extendable and is the most mature way to do recommendation. You can choose different combinations of features and different form of interactions. For eCommerce, for example, the interaction could be glance view, add to cart, purchase etc. For social media, the interaction could be follow, friends etc.  For a new user or new item, once we get their features, we can make a prediction. Cold start problem is solved, and we can add user and item features freely.

## Step 1: Building a QA bot
In fact, we are doing QA hundreds even thousands times everyday, on Google or ChatGPT. They represent two main-stream pathway to building a QA bot: <font color='blue'>Retrieval-based</font> and <font color='blue'>Large Language Model-based</font>. They have fundamentally different way of retrieving and synthesizing data, thus totally different way to develope. 

### Retrieval-based QA bot

```python
import utils
utils.show_img('../NLP/img/qa-retrieval.png', size_inch=18)
```

Reading material: https://arxiv.org/abs/1704.00051 by Danqi Chen. This paper gives a pretty solution of retrieval-based QA system. Construction of Retrieval-based QA bot usually have two steps:

a) Retrieving topK most relevant doc from the database. Sounds so familiar, right? This is a search engine we introduced in previous section. You could use any embedding method like TFIDF, GloVes, encoder or model based embedding (RNN, Bert) and any kind of relevance score like cosine similarity or BM25. Or you could train a model to predict the relevance score based on the embedding. You are free to use method like hierachical topics to find the topK. 

b) Build a model to find the selected relevant docs' most relevant part. The K docs retrieved in previous steps are concatenated and a finer model is trained to predict the starting and ending point of an answer for a given query. 

The biggest benefit of using retrieval-based system is that you have full control of the model and you have full data privacy protection. The system is mature, cheap and light-weighted. Besides, the model is fully explanable. If you care about data privacy and explanability, this is the way to go.

### Large Language Model - based QA bot
A Large Language Model is doing a completion task. Given some prompts and context, the model is expected to fill the remaining part. It is fundamentally a statistical model that choose the most likely token (word) one by one. Contrary to retrieval-based model, LLM does not have a database. Instead, it memorize all the information in the model parameters (https://arxiv.org/abs/2002.08910). Thus no doubt, bigger model better performance. All the information retrieval process happen inside the model. Furthermore, the model does synthesis, and will directly output the answer to a query, just as you see with ChatGPT. Good thing is that the synthesis is really helpful, you don't need to filter the relevant docs yourself. The bad thing is:

    a) we don't know why we have this answer... 

    b) It is hard to notice if something goes wrong. (Recall the big mistake about JWST Google's Bard made during its first demo. Also, the hallucination of all generative models including chatGPT).

If you want to develop a LLM-based QA bot with some domain knowledge, or use it for some adapted tasks, you have several options:

If data privacy is a problem (For example, Amazon does not allow you to send data to OpenAI. Many companies have similar data privacy regularizations.), you must use an open source model. 

    1) Use open source model like GPT2, Bert, T5 and Bart, GPT-J, LLaMA and fine tune them using your own knowledge base. (Tsinghua: https://arxiv.org/abs/1905.07129 PKU: https://arxiv.org/abs/2010.08824 ) The knowledge base must be created first and transformed into target tasks to complete the training. One way to create high-quality QA dataset is to get response from OpenAI API. You can use this dataset to train your own open-source QA model. But make sure you comply with OpenAI policy on generated data usage. It forbidden using geneated data to develop commercial competing fundation models. 

    2) Use open source model and add adapters (https://arxiv.org/abs/2212.08120). This is another kind of fine-tuning, but you are fine-tuning the adapter instead of the whole model thus cheaper relative to method 1). 

    3) Use open source models and add memory in a framework like LangChain, GPT-index. This is cheaper than fine-tuning. 

If you are free to use external API like OpenAI API:
    4) Use OpenAI API and fine tune the model. (https://platform.openai.com/docs/guides/fine-tuning) Note, this solution is super expensive because you pay for fine-tuning the model and you pay 7 times normal API call rate for calling the API. 

    5) Add a topK selector yourself, send the context in prompt, and use API to do the sysnthesis. https://gpt-index.readthedocs.io/en/latest/index.html https://langchain.readthedocs.io/en/latest/
    One example of building QA KnowledgeBot is here: https://www.youtube.com/watch?v=Dhc_fq5iCnU .
