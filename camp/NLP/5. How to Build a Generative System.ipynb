# How to Finetune your LLM

Step-by-Step Instructions
1. Two ways of finetuning: Supervised Fine-Tuning(SFT) and RLHF. They have different goals, thus different data format, and different tools to use. 

Instructions: https://github.com/yaodongC/awesome-instruction-dataset
RLHF: https://huggingface.co/datasets/Anthropic/hh-rlhf

2. Choose your LLM based on your need considering license, performance, language, size of model, etc. 

https://huggingface.co/spaces/HuggingFaceH4/open_llm_leaderboard

3. Decide which way you want to go: full model distributed training vs peft (e.g. lora, qlora). 

4. Find the resources to finetune your model. This is the most chaotic part, you will struggle, both technically and emotionally :-P This is an active area, no established path. 

5. Deployment 

Example: Finetuning Falcon-40B on Sagemaker

https://github.com/maekawataiki/aws-ml-jp/blob/feat/falcon/tasks/generative-ai/text-to-text/fine-tuning/instruction-tuning/Transformers/Falcon_LoRA.ipynb
