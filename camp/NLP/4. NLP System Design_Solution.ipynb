<font color='red'>Question:</font> What is the difference precise topK retrieval and approximate nearest neighbor retrieval? 

<font color='red'>A:</font>  
Precise Top-K Retrieval and Approximate Nearest Neighbor (ANN) Retrieval are two different approaches to searching for similar items or data points in a large dataset. Here are the key differences between the two:

Accuracy vs. Efficiency:

Precise Top-K Retrieval: In this approach, the goal is to find the exact top K nearest neighbors to a given query point. It focuses on accuracy and guarantees that the returned results are the true nearest neighbors. This method is used when exact matches are crucial, but it can be computationally expensive for large datasets.

Approximate Nearest Neighbor Retrieval: ANN retrieval, on the other hand, sacrifices some degree of accuracy to gain efficiency. It aims to find an approximate set of nearest neighbors that are very close to the true nearest neighbors but not necessarily identical. ANN methods are designed to be much faster for high-dimensional data and are suitable for scenarios where getting exact matches is less critical.

Use Cases:

Precise Top-K Retrieval: This is often used in applications where precision and correctness are paramount, such as database search, cryptography, and certain machine learning tasks. For example, when searching for duplicate records in a database, you want to ensure that the retrieved results are exact matches.

Approximate Nearest Neighbor Retrieval: ANN methods are employed in scenarios where speed is a priority, and a small trade-off in accuracy is acceptable. Common applications include image retrieval, recommendation systems, and information retrieval in large-scale databases. ANN methods are particularly effective in high-dimensional spaces, such as text and image data.

Computational Complexity:

Precise Top-K Retrieval: This approach often involves exhaustive searches, which can be computationally expensive, especially as the dataset size grows. It typically has a time complexity of O(N), where N is the number of data points in the dataset.

Approximate Nearest Neighbor Retrieval: ANN methods leverage techniques like locality-sensitive hashing (LSH) and tree-based indexing structures to significantly reduce the number of distance computations, making them more efficient, especially in high-dimensional spaces. They often have sublinear or logarithmic time complexity, such as O(log(N)).

Scalability:

Precise Top-K Retrieval: It can become impractical for very large datasets due to the computational cost of exhaustively comparing the query point with all data points.

Approximate Nearest Neighbor Retrieval: ANN methods are designed to be scalable and are well-suited for handling large datasets efficiently.

<font color='red'>Question:</font> Is it possible to enhance the performance of precise topK retrieval? How?

【 A. 1. When K<<N, Use Heap to reduce time complexity from NLogN in sorting to NLogK

```python
import heapq
import numpy as np

def precise_top_k_retrieval(query_point, dataset, k):

    # Initialize a min heap to store distances and their corresponding indices
    min_heap = []

    # Iterate through the dataset and calculate distances
    for i, data_point in enumerate(dataset):
        distance = np.linalg.norm(query_point - data_point)  # Euclidean distance
        # Push the negative distance and index onto the heap
        heapq.heappush(min_heap, (-distance, i))
        
        # If the heap size exceeds K, pop the smallest element
        if len(min_heap) > k:
            heapq.heappop(min_heap)

    # Extract the indices of the top K nearest neighbors
    top_k_neighbors = [index for _, index in min_heap]
    # Reverse the list to get the correct order (smallest distance to largest)
    top_k_neighbors.reverse()
    
    return top_k_neighbors

top_k_neighbors = precise_top_k_retrieval(query_point, dataset, K)
print("Indices of top K nearest neighbors:", top_k_neighbors)
```

<font color='red'>Question:</font> Explain the LSH algorithm. Why making an inverted index table? Why this is much faster? 

【A:

The essential idea of FAISS is how to gain high search speed if we allow to sacrifice some accuracy. (tech evolve: KNN -> ANN -> FAISS)

LSH is Locality-Sensitive-Hashing. The original high-dimensional vectors are projected to a low-dimensional space and become hash codes. The samples clustered in the high dimension are supposed to be projected in a same hash code "bucket". You can think the overall hash code of the bucket as the centroid of a cluster. When a query comes, it is projected into the same low dimensional space and obtain its own hash code. Comparing the query's hash code to the bucket hash codes, FAISS quickly knows which bucket the query belongs to, and then it performs a brutal search in that bucket. LSH algorithm greatly reduces the searching scope and also reduces the vector dimension, which makes searching much faster.

A normal forward index system maps each document to the terms (features) it contains, while an "inverted" index system maps from terms (features) in the dataset to the document that contains the terms. This is like to reverse the keys and values of a dictionary. In searching, we compare vectors in a query to the existing samples, so it's comparing terms and outputting the most similar document(s). Therefore, an inverted index (vectors -> doc) works much faster than a forward index (doc -> vectors).

Hierarchical Navigable Small World (HNSW) overview:

Key Concepts:

1. Graph Structure:
HNSW constructs a graph where each node represents a data point. Nodes are connected to each other based on their proximity in the high-dimensional space.

2. Hierarchical Structure:
The graph is organized in multiple layers, each representing a different level of proximity or granularity. The layers are interconnected to form a hierarchy.

3. Navigability:
HNSW aims to maintain a "navigable small-world" property, where most nodes are connected to other nodes that are both close in the high-dimensional space and easily reachable via a small number of hops.

Algorithm Steps:

1. Construction:

Layer 0: Build the initial layer (base layer) of the graph by connecting nodes in the dataset based on their distances in the high-dimensional space.

Hierarchical Construction: For each subsequent layer, use a heuristic to connect nodes to each other based on their proximity. The higher layers have fewer connections, creating a hierarchy of nodes.

2. Search:

Start the search from the highest layer (coarsest granularity) and progressively move to lower layers, aiming to reach the approximate nearest neighbor efficiently.

Traverse the graph layer by layer, navigating toward nodes that are closer to the target in the high-dimensional space.

Recommendation System is fundamentally an Information Retrieval system. Let me explain why. 

### Brutal Force

Think about a brutal force collaborative filtering method for user-based or item-based recommendations. Collaborative filtering is commonly used for recommender systems. These techniques aim to fill in the missing entries of a user-item association matrix.

User-based method is that you find top k users most similar to your target user, and average their behavior (rating, purchase, like/dislke). Then you can predict what are the most likely items he/she will purchase/like.

Item-based method is reverted. You find top k items that most similar to your target item, and average their relavent history (rating, purchase, like/dislike). Then you can pick top n people who are most likely to engage in the interaction (purchase, like, watch) and do your marketing to them. 

Sounds so similar to search engine, right? Similarity, top k... Because this is fundamentally the same matching game of two sides. 

【Collaborative filtering's logic: if two users A and B have similar preference, the items that A likes can be recommended to B; if two items A and B are similar, we can recommend B to the users who have purchased A. 

Matrix factorization is a model-based approach to achieve collaborative filtering (identify the similarity between users and between items, then give recommendation). Matrix factorization is able to extract complex and implicit relationship from user-item interaction matrix, but it can't use important information outside the user-item interaction matrix.

Using features of users, items and their interaction is a feature-based approach to achieve collaborative filtering. This approach is able to utilize important features outside user-item interaction matrix, but it may miss useful some potential implicit features.

Since these two approaches are complementary to each other, the third approach is a hybrid: after making both models, combine the models by weighted average, cascade or switching.】

The link of retrieval system with matrix factorization method is less straightforward, but let me explain briefly. You have a matrix containing the interaction (rating, purchase, like) between customers and items (movie, iPhone etc). Some elements in the matrix is missing, and we are supposed to fill those missing values. To achieve this, you can approximate the interacton matrix with a multiplication of two matrices: user matrix and item matrix. And the full matrix prediction can be obtained by multiplication of the two matrices.

<font color='red'>Question:</font> What is the problem of collaborative filtering method?

【The problem of brute force collaborative filtering method is essentially the same as brutal search, scalability. Brute force collaborative filtering method calculates the similarity between every possible pair of users and items in a databse. When the database is huge, the computation becomes too expensive.

Collaborative filtering suffers from several challenges, including:

Cold-start problem: Collaborative filtering requires data on user-item interactions to make recommendations. In cases where a new user or item enters the system, there may not be enough data to make accurate predictions.

Sparsity: Collaborative filtering can become less effective as the number of users and items in the system grows. In many real-world scenarios, user-item interactions are sparse, meaning most users have interacted with only a small fraction of the items in the system.

Scalability: Collaborative filtering can become computationally expensive as the number of users and items in the system grows. To make predictions for a new user or item, the algorithm needs to compare the user or item with all other users or items in the system.

Gray sheep problem: In some cases, a user may have a unique set of preferences that do not align with any other user in the system. This user is referred to as a "gray sheep" and may not receive accurate recommendations.

Shilling attacks: Collaborative filtering is vulnerable to "shilling attacks," where malicious users create fake profiles or artificially inflate their interactions to manipulate the system's recommendations.

Researchers are working on addressing these challenges with new techniques and algorithms, such as hybrid recommendation systems, content-based filtering, and deep learning models.

loss = Σ($R_{ij}$-$P_{ik}$•$Q_{kj}$)^2 + β(Σ$P_{ik}$^2 + Σ$Q_{kj}$^2)/2, where β is the weight for L2 term

gradient: ∂loss/∂$P_{ik}$ = -2($R_{ij}$-$P_{ik}$•$Q_{kj}$)$Q_{kj}$ + β*$P_{ik}$, and

∂loss/∂$Q_{kj}$ = -2($R_{ij}$-$P_{ik}$•$Q_{kj}$)$P_{ik}$ + β*$Q_{kj}$,

updated weight ($P_{ik}$ or $Q_{kj}$) = original weight - α * gradient, where α is the learning rate

nice work!

```python
import numpy

def matrix_factorization(R, P, Q, K, steps=5000, alpha=0.0002, beta=0.02):
    '''
    R: rating matrix
    P: |U| * K (User features matrix)
    Q: |D| * K (Item features matrix)
    K: latent features
    steps: iterations
    alpha: learning rate
    beta: regularization parameter'''
    Q = Q.T

    for step in range(steps):
        for i in range(len(R)):
            for j in range(len(R[i])):
                if R[i][j] > 0: #only use truly rated elements in R to update P&Q
                    # calculate RMSE error
                    eij = R[i][j] - numpy.dot(P[i,:],Q[:,j])

                    for k in range(K):
                        # update weights (P,Q) by gradients: -alpha(∂loss/∂𝑃) or -alpha(∂loss/∂Q)
                        P[i][k] = P[i][k] + alpha * (2 * eij * Q[k][j] - beta * P[i][k])
                        Q[k][j] = Q[k][j] + alpha * (2 * eij * P[i][k] - beta * Q[k][j])

        # calculate the loss again using updated P and Q
        R_pred = numpy.dot(P,Q)
        total_error = 0

        for i in range(len(R)):

            for j in range(len(R[i])):

                if R[i][j] > 0:
                    # add MSE
                    total_error = total_error + pow(R[i][j] - R_pred[i,j], 2)
                    # add L2 penalty
                    for k in range(K):
                        total_error = total_error + (beta/2) * (pow(P[i][k],2) + pow(Q[k][j],2))
        
        # if loss is very small stop updating, otherwise update "steps" times
        if total_error < 0.001: # here the author didn't consider average error
            break

    return P, Q.T
```

Denise Chen's matrix_factorization() is nicely coded, but some parts are repetitive. I further simplified the code as follows：

```python
def simplify_MF(R, K, steps=5000, alpha=0.0002, beta=0.02):
    '''
    R: user-item rating matrix, array type
    K: latent features
    steps: iterations
    alpha: learning rate
    beta: regularization parameter
    return: user-feature matrix P and feature-item matrix Q.T
    '''
    
    N,M = R.shape
    P = np.random.rand(N,K)
    Q = np.random.rand(K,M)   
    error = R - numpy.dot(P,Q) #initial error matrix
    num_ratings = (R>0).sum() # number of true ratings
    
    for step in range(steps):
        total_loss = 0
        for i in range(len(R)):
            for j in range(len(R[i])):
                if R[i][j] > 0: #only use truly rated elements to update P & Q and to calculate loss                 
                    
                    total_loss += error[i][j]**2 #add MSE
                    
                    for k in range(K):
                        # update weights (P,Q) by gradients: -alpha(∂loss/∂𝑃) or -alpha(∂loss/∂Q)
                        P[i][k] = P[i][k] + alpha * (2 * error[i][j] * Q[k][j] - beta * P[i][k])
                        Q[k][j] = Q[k][j] + alpha * (2 * error[i][j] * P[i][k] - beta * Q[k][j])
                    
                        total_loss += (beta/2) * (P[i][k]**2 + Q[k][j]**2) # add L2 penalty
        
        # update error matrix
        error = R - numpy.dot(P,Q)

        if total_loss / num_ratings < 0.00005: # set threshold on average loss
            break

    return P, Q.T
```

### Transforming recommendation problem into a regression or classification problem

By answering the above questions you probably have realized, <font color='blue'>dealing with interaction matrix is powerful but limited.</font> The latent parameters carry some information of users and items, but why we don't use user and item features directly? Thus we need to wait for them to accumulate some data and re-train the model. How can we use the features of users and items and solve the cold start problem at the same time?

We can <font color='blue'>convert the recommendation problem into a regression or classification problem by using features to predict the score or interaction.</font> This design logic is super powerful and extendable and is the most mature way to do recommendation. ### Large Language Model - based QA bot
A Large Language Model is doing a completion task. Given some prompts and context, the model is expected to fill the remaining part. It is fundamentally a statistical model that choose the most likely token (word) one by one. Contrary to retrieval-based model, LLM does not have a database. Instead, it memorize all the information in the model parameters (https://arxiv.org/abs/2002.08910 use T5 to answer open domain question, close-book exam). Thus no doubt, bigger model better performance. All the information retrieval process happen inside the model. Furthermore, the model does synthesis, and will directly output the answer to a query, just as you see with ChatGPT. Good thing is that the synthesis is really helpful, you don't need to filter the relevant docs yourself. The bad thing is:

    a) we don't know why we have this answer... 

    b) It is hard to notice if something goes wrong. (Recall the big mistake about JWST Google's Bard made during its first demo. Also, the hallucination of all generative models including chatGPT).
